## COVID19 AND MENTAL HEALTH: Interactive Dashboard

## Considering the impacts of COVID-19, how is Singapore's mental health status?

```python
import pandas as pd

path = '/mnt/c/users/qamar/Documents/4 Development/Jupyter/mindful_dashboarding/COVID-19-master/csse_covid_19_data/csse_covid_19_time_series/covid19_ALL_global.csv'
df = pd.read_csv(path)
df = df.drop(df.columns[9:],axis=1)
```

```python
## REMOVE LAT/LONG NaN
df_location = df.dropna(subset=['lat','long'])

## REGION DF
df_region = df['region']

## COUNTRY/REGION NAME DF
df_country = df['country']

## CONFIRMED DF
df_confirmed = df['confirmed']
sum_confirmed = df.iloc[1]['value']

## DEATHS DF
df_deaths = df['deaths']
sum_deaths = df.iloc[2]['value']

## RECOVERED DF
df_recovered = df['recovered']
sum_recovered = df.iloc[3]['value']

## LOCATION DF
df_geo = df[['lat','long']]
```

```python
import folium
f = folium.Figure(height=400)
m = folium.Map(location=[48, -102], zoom_start=2, min_zoom = 2, max_zoom = 6,tiles="cartodbpositron").add_to(f)


for i in range(len(df_geo)):
    folium.Circle(
        location = [df_geo['lat'][i],df_geo['long'][i]],
        name="choropleth",
        data=df,
        fill=True,
        radius=200000,
        color=False,
        fill_color='red',
        legend_name="Number of Confirmed Cases",
        tooltip = '<h5>'+str(df_region.iloc[i])+', '+df_country.iloc[i]+'</h5>'+
        '<li style="list-style-type: none">Confirmed: '+str(df_confirmed.iloc[i])+'</li>'+
        '<li style="list-style-type: none">Recovered: '+str(df_recovered.iloc[i])+'</li>'+
        '<li style="list-style-type: none">Deaths: '+str(df_deaths.iloc[i])+'</li>'
    ).add_to(m)

    folium.LayerControl().add_to(m)

m
```
